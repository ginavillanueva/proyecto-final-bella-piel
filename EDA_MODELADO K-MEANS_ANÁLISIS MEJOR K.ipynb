MODELADO "K-MEANS"

Para llevar a cabo el desarrollo de algoritmos de clustering, como K-Means, en Python dentro de Jupyter Lab y cumplir con el objetivo de segmentar a los formuladores en grupos con características similares, sigue estos pasos:

```python
# Importar bibliotecas necesarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Seleccionar las variables relevantes para el clustering
variables_clustering = ['CANTIDAD', 'COMISION']
X = df[variables_clustering]
```

```python
# Normalizar los datos para asegurar que todas las variables tengan la misma escala
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Determinar el número óptimo de clusters utilizando el método del codo (Elbow Method)
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Visualizar el método del codo para seleccionar el número óptimo de clusters
plt.figure(figsize=(10, 6))
```

```python
# Visualizar el método del codo para seleccionar el número óptimo de clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método del Codo para K-Means Clustering')
plt.xlabel('Número de Clusters')
plt.ylabel('Inertia')
plt.show()
```

Este fragmento de código permite visualizar el método del codo y determinar el número óptimo de clusters observando el punto donde la inercia deja de disminuir rápidamente (el "codo" en el gráfico). 
Basado en el gráfico, un buen candidato para el número de clusters sería 3, ya que es en este punto donde el gráfico comienza a aplanarse y la inercia disminuye a un ritmo más lento.
Elegir más de 3 clusters no parece proporcionar una ganancia significativa en términos de disminuir la inercia y puede llevar a un sobreajuste, donde los clusters adicionales pueden no tener significado práctico o estadístico.
Mas cabe recalcar que tambien podria verse usar un k=4 debido a que tambien es un punto critico en la ggrafica donde el gráfico comienza a aplanarse y la inercia disminuye a un ritmo más lento. por lo cual se analizará en ambos puntos y se eligira el mejor.

CON K=3 APLICANDO K.MEANS SE TIENE

```python
# Aplicar K-Means con el número óptimo de clusters k=3
num_clusters_optimo = 3  
kmeans_optimo_k_3 = KMeans(n_clusters=num_clusters_optimo, random_state=42)
df['Cluster'] = kmeans_optimo.fit_predict(X_scaled)
```

CON K=4 APLICANDO K.MEANS SE TIENE

```python
# Aplicar K-Means con el número óptimo de clusters k=4
num_clusters_optimo = 4 
kmeans_optimo_k_4 = KMeans(n_clusters=num_clusters_optimo, random_state=42)
df['Cluster'] = kmeans_optimo.fit_predict(X_scaled)
```

Análisis del mejor modelo con "k"

```python
from sklearn.metrics import silhouette_score

# Ajustar K-Means con k=3 y obtener las etiquetas
kmeans_optimo_k_3 = KMeans(n_clusters=3, random_state=42)
kmeans_optimo_k_3.fit(X_scaled)
labels_k3 = kmeans_optimo_k_3.labels_

# Ajustar K-Means con k=4 y obtener las etiquetas
kmeans_optimo_k_4 = KMeans(n_clusters=4, random_state=42)
kmeans_optimo_k_4.fit(X_scaled)
labels_k4 = kmeans_optimo_k_4.labels_

# Calcular el Coeficiente de Silueta para k=3 y k=4
silhouette_k3 = silhouette_score(X_scaled, labels_k3)
silhouette_k4 = silhouette_score(X_scaled, labels_k4)

print("Silhouette Coefficient for k=3: ", silhouette_k3)
print("Silhouette Coefficient for k=4: ", silhouette_k4)

```

En resumen, aunque k=3 tiene un coeficiente de silueta ligeramente superior y podría ser considerado el mejor en términos de métricas de agrupamiento puro, la decisión final también debe basarse en la relevancia práctica y la interpretación de los clusters.

TIPO_COMPRA

```python
from scipy.stats import kruskal

# Prueba de Kruskal-Wallis para la variable 'Cantidad'
for cluster in range(num_clusters_optimo):
    cluster_df = df[df['Cluster'] == cluster]
    unique_tipos_compra = cluster_df['TIPO_COMPRA'].nunique()
    
    if unique_tipos_compra >= 2:
        print(f"\nPrueba de Kruskal-Wallis para CANTIDAD en Cluster {cluster}:")
        print(kruskal(*[group['CANTIDAD'] for name, group in cluster_df.groupby('TIPO_COMPRA')]))
    else:
        print(f"\nNo se pueden realizar pruebas de Kruskal-Wallis en Cluster {cluster} debido a la falta de categorías únicas en 'TIPO_COMPRA'.")


```
