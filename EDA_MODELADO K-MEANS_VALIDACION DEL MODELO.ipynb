MODELADO "K-MEANS"

Para llevar a cabo el desarrollo de algoritmos de clustering, como K-Means, en Python dentro de Jupyter Lab y cumplir con el objetivo de segmentar a los formuladores en grupos con características similares, sigue estos pasos:

```python
# Importar bibliotecas necesarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Seleccionar las variables relevantes para el clustering
variables_clustering = ['CANTIDAD', 'COMISION']
X = df[variables_clustering]
```

```python
# Normalizar los datos para asegurar que todas las variables tengan la misma escala
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Determinar el número óptimo de clusters utilizando el método del codo (Elbow Method)
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Visualizar el método del codo para seleccionar el número óptimo de clusters
plt.figure(figsize=(10, 6))
```

```python
# Visualizar el método del codo para seleccionar el número óptimo de clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método del Codo para K-Means Clustering')
plt.xlabel('Número de Clusters')
plt.ylabel('Inertia')
plt.show()
```

Este fragmento de código permite visualizar el método del codo y determinar el número óptimo de clusters observando el punto donde la inercia deja de disminuir rápidamente (el "codo" en el gráfico). 
Basado en el gráfico, un buen candidato para el número de clusters sería 3, ya que es en este punto donde el gráfico comienza a aplanarse y la inercia disminuye a un ritmo más lento.
Elegir más de 3 clusters no parece proporcionar una ganancia significativa en términos de disminuir la inercia y puede llevar a un sobreajuste, donde los clusters adicionales pueden no tener significado práctico o estadístico.

```python
# Aplicar K-Means con el número óptimo de clusters
num_clusters_optimo = 3  
kmeans_optimo = KMeans(n_clusters=num_clusters_optimo, random_state=42)
df['Cluster'] = kmeans_optimo.fit_predict(X_scaled)
```

```python
# Visualizar los clusters en el espacio bidimensional
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(12, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Cluster'], cmap='viridis', edgecolors='k', alpha=0.7)
plt.title('Resultados del Clustering (K-Means)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
# Analizar las características de cada cluster
cluster_analysis = df.groupby('Cluster')[variables_clustering].mean()
print(cluster_analysis)
```

-	El Clúster 0 tiene valores medios negativos para 'CANTIDAD' y 'COMISIÓN', lo cual, después de la normalización, sugiere que este grupo puede representar a los formuladores que tienen un rendimiento por debajo de la media en términos de la cantidad de productos prescritos y las comisiones generadas.
-	El Clúster 1 tiene valores medios significativamente altos tanto para 'CANTIDAD' como para 'COMISIÓN', lo que indica que este grupo podría estar compuesto por formuladores de alto rendimiento que prescriben en grandes cantidades y, posiblemente, generan altas comisiones.
-	El Clúster 2 tiene valores medios más moderados, pero aún por encima del promedio, sugiriendo que este grupo podría representar a los formuladores con un rendimiento bueno, pero no tan extremo como el Cluster 1.

```python
cluster_0_df = df[df['Cluster'] == 0]
cluster_1_df = df[df['Cluster'] == 1]
cluster_2_df = df[df['Cluster'] == 2]
```

Prueba de Kruskal-Wallis:

```python
from scipy.stats import kruskal

# Ejemplo para la variable 'CANTIDAD'
result_kruskal_cantidad = kruskal(cluster_0_df['CANTIDAD'], cluster_1_df['CANTIDAD'], cluster_2_df['CANTIDAD'])
print("\nPrueba de Kruskal-Wallis para CANTIDAD:")
print(result_kruskal_cantidad)

# Ejemplo para la variable 'COMISION'
result_kruskal_comision = kruskal(cluster_0_df['COMISION'], cluster_1_df['COMISION'], cluster_2_df['COMISION'])
print("\nPrueba de Kruskal-Wallis para COMISION:")
print(result_kruskal_comision)
```

Análisis de Varianza (ANOVA):

```python
from scipy.stats import f_oneway

# Ejemplo para la variable 'CANTIDAD'
result_anova_cantidad = f_oneway(cluster_0_df['CANTIDAD'], cluster_1_df['CANTIDAD'], cluster_2_df['CANTIDAD'])
print("\nAnálisis de Varianza (ANOVA) para CANTIDAD:")
print(result_anova_cantidad)

# Ejemplo para la variable 'COMISION'
result_anova_comision = f_oneway(cluster_0_df['COMISION'], cluster_1_df['COMISION'], cluster_2_df['COMISION'])
print("\nAnálisis de Varianza (ANOVA) para COMISION:")
print(result_anova_comision)
```

Comparaciones Post-hoc:

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Ejemplo para la variable 'CANTIDAD'
tukey_cantidad = pairwise_tukeyhsd(df['CANTIDAD'], df['Cluster'])
print("\nComparación Post-hoc (Tukey) para CANTIDAD:")
print(tukey_cantidad)

# Ejemplo para la variable 'COMISION'
tukey_comision = pairwise_tukeyhsd(df['COMISION'], df['Cluster'])
print("\nComparación Post-hoc (Tukey) para COMISION:")
print(tukey_comision)
```
